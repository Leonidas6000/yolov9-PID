### Solo es necesario modificar esta linea directory = f"{HOME}/yolov9-PID/runs/detect/exp" y al final dejar el exp como esta o ponerle 1 o 2 o el número que corresponda, solo mirando el resultado del comando anterior podemos ver en que exp ha metido los resultados.

```python
from IPython.display import Image, display

# Directorio donde se encuentran las imágenes
directory = f"{HOME}/yolov9-PID/runs/detect/exp2"

# Obtener una lista de nombres de archivo en el directorio
image_files = [f for f in os.listdir(directory) if f.endswith(('.jpg', '.jpeg', '.png', '.gif'))]

# Mostrar cada imagen
for filename in image_files:
    image_path = os.path.join(directory, filename)
    display(Image(filename=image_path, width=600))
```
